```python
from ultralytics import YOLO
```

```python
model_path = fr"runs\detect\train5\weights\last.pt"
yaml_path = "datasets/vaca/data.yaml"
image_path = r"datasets\vaca\images\val\vj32_jpg.rf.26e063f5a917a9e05dedcb26ab7726d1.jpg"
```

```python
from pathlib import Path
import hashlib

p = Path(model_path)
print("exists:", p.exists())
print("size bytes:", p.stat().st_size if p.exists() else None)

# hash pra ver se muda entre downloads (se mudar, tÃ¡ instÃ¡vel/corrompendo)
if p.exists():
    h = hashlib.sha256(p.read_bytes()).hexdigest()
    print("sha256:", h[:32], "...")
```

```python
model = YOLO(model_path)
```

```python
print(model.names)
```

```python
print(model.ckpt['train_args'].get("epochs"))
```

```python
metrics = model.val(data=yaml_path)
print("")
print(50*"-")
print("Validation Metrics:")
print("mAP50:", metrics.box.map50.round(3))
print("mAP50-95:", metrics.box.map.round(3))
print("Precision:", metrics.box.mp.round(3))
print("Recall:", metrics.box.mr.round(3))
```

```python
results = model(image_path)
```
